---
<h2>1. Importing the Dataset</h2>

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../../../../data/clean/Position_Salaries.csv')
display(df.head())
x = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

---
<h2>2. Training the Model with the Whole Dataset</h2>

```python
from sklearn.tree import DecisionTreeRegressor

'''This algorithm works well in high dimensional features!!'''
dtr = DecisionTreeRegressor(random_state=0)
dtr.fit(x, y)
```

---
<h2>3. Visualizing the Model Results</h2>

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('seaborn-whitegrid')
%matplotlib inline

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6,4))

regression_x = np.arange(x.min(), x.max()+1, 0.1)
regression_y = [dtr.predict([[i]])[0] for i in regression_x]

ax.scatter(x, y, color='red')
ax.plot(regression_x, regression_y, color='blue')
ax.set_xlabel("Position level")
ax.set_ylabel("Salary appx")
ax.set_title("benchmark average salary by position level")

plt.show()
```

---
<h2>4. Prediction of Possible Salary based on Position Level as Input</h2>

```python
post_lvl = 6.5
y_pred = dtr.predict([[post_lvl]])

print("The salary estimate for for position level at {} is ${:,.2f}".format(post_lvl, y_pred[0]))
```
