<p style="font-size:13px">Probability Distribution is a statistical function which links or lists all the possible outcomes a random variable can take, in any random process, with its corresponding probability of occurrence. </p>

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.style.use('seaborn-whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('../../data/clean/house_data.csv',
                 usecols=['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'condition', 'grade', 'sqft_above', 'sqft_basement'])
df.drop_duplicates(inplace=True)
df.dropna(axis=0, inplace=True)
df = df[df.bedrooms != 0]
df.head()
```

---
<h1>CDF</h1>
<ul style="font-size:13px">
    <li><i>Cumulative distribution function</i> (<strong>CDF</strong>) will give you the probability that a random variable is less than or equal to a certain real number<ul>
            <li><strong><i>Cumulative</i></strong> — to accumulate; to increase in quantity over time
            <li><strong><i>Distribution</i></strong> — the way in which a quantity (or probability) is spreadout (is it Left-Skewed, Right-Skewed or Symetric distribution)
            <li><strong><i>Function</i></strong> — tells us the output of given input
        </ul>
    <li>CDF is applicable for describing the distribution of random variables either it is continuous or discrete<ul>
            <li><strong><i>PDF</i></strong> shows the probability of one, specific x-value occuring
            <li><strong><i>CDF</i></strong> shows the probability of all x-values up to a certain x occuring
        </ul>
    <li>The value of CDF is start from 0 and 1 as the maximum value so we can never have a negative probability (under 0) of event occuring and the probability value never exceed 1
</ul>

<h1>PDF</h1>
<img src="../../images/pdf.JPG" style="background:white; display: block; margin-left: auto;margin-right: auto; width:60%"/>
<ul style="font-size:13px">
    <li><i>Probability density function</i> (<strong>PDF</strong>) of a random variable X, when integrated over a set of real numbers A, will give the probability that X lies in A
    <li>Its used for a continuous random variable<ul>
            <li>Continuous random variable is a variable which having the values between the range/interval and take infinite number of possible ways
            <li>the variables whose values are obtained by measuring is called Continuous random variable
        </ul>
    <li>On PDF graph the probability of single outcome is <strong>always zero</strong>, this happened because the single point represents the line which doesn’t cover the area under the curve
    <li>Distribution types<ul>
            <li><strong>Exponential Distribution</strong> — It is used to model the probability distribution of the time between processes in which events occur continuously and independently at a constant average rate <i><strong>λ</strong></i> like the amount of time from now until an earthquake occurs
            <li><strong>Continuous Uniform Distribution</strong> — It describes an experiment where the outcome is continuous which can take any value between a minimum and maximum value denoted as <i><strong>a</strong></i> and <i><strong>b</strong></i>
            <li><strong>Gaussian Distribution</strong> — This distribution is used to represent any real valued random variable whose distribution is unknown where Mean <i><strong>μ</strong></i> and Variance <i><strong>σ²</strong></i> are used to denote it
        </ul>
    <li>ie. : a population of women heights
</ul>

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,4))

sns.distplot(df['sqft_living'], kde=True, color='skyblue', ax=ax[0])
ax[0].set_xlabel('size of livingroom in squarefeet')
ax[0].set_ylabel('Frequency')

sns.distplot(df['sqft_living'], hist_kws={'cumulative':True, 'density': True}, kde_kws={'cumulative':True}, color='skyblue', ax=ax[1])
ax[1].set_xlabel('size of livingroom in squarefeet')
ax[1].set_ylabel('% percentage of cumulative')

plt.suptitle('PDF and CDF')
plt.show()
```

<h1>PMF</h1>
<img src="../../images/pmf.JPG" style="background:white; display: block; margin-left: auto;margin-right: auto; width:60%"/>
<ul style="font-size:13px">
    <li><i>Probability mass function</i> (<strong>PMF</strong>) gives you the probability that a discrete random variable is exactly equal to some real value
    <li>It is used to assign the probability of a discrete random variable<ul>
            <li>Discrete Random Variable is a variable which takes countable number of distinct values
            <li>Or the variables whose values are obtained by counting
        </ul>
    <li>Distribution types<ul>
            <li><strong>Bernoulli Distribution</strong> — It can be thought of a model for the experiment whose outcome is binary<ul>
                    <li>ie. : head/tail, yes/no, etc
                </ul>
            <li><strong>Binomial Distribution</strong> — This distribution with parameter n and p is the generalized case of Bernoulli distribution where <i><strong>n</strong></i> is the number of sequence of random experiment and <i><strong>p</strong></i> is the probability of a positive outcome in each experiment<ul>
                    <li>ie. : tossing a coin 20 times as <i><strong>n</strong></i> and probability of getting 9 heads as <i><strong>p</strong></i>
                </ul>
            <li><strong>Geometric Distribution</strong> — This type of distribution is to model the number of Bernoulli trials with probability <i><strong>p</strong></i> of success until first success<ul>
                    <li>ie. : throwing a die multiple times until ‘1’ appears for the first time
                </ul>
            <li><strong>Poisson Distribution</strong> — This distribution is to model the probability of a given number of events <i><strong>k</strong></i> occurring in a fixed interval of time or space if these events occur with a known constant mean rate <i><strong>λ</strong></i> and independently of the time since the last event<ul>
                    <li>ie. : the number of phone calls received by a call center or the number of emails per day or the number of users visited on a website in an interval
                </ul>
            <li><strong>Discrete Uniform Distribution</strong> — It is used to model the random experiments where all the discrete outcomes between <i><strong>a</strong></i> and <i><strong>b</strong></i> (<i><strong>b ≥ a</strong></i>) are equally likely like throwing a fair die
        </ul>
</ul>

```python
from empiricaldist import Pmf, Cdf

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,4))

sns.distplot(Pmf(df['condition']), kde=False, color='skyblue', ax=ax[0])
ax[0].set_xlabel('rate of house\'s condition')
ax[0].set_ylabel('Frequency')

sns.distplot(Cdf(df['condition']), hist_kws={'cumulative':True}, kde_kws={'cumulative':True}, color='skyblue', ax=ax[1])
ax[1].set_xlabel('rate of house\'s condition')
ax[1].set_ylabel('% percentage of cumulative')

plt.suptitle('PMF and CDF')
plt.show()
```
